```python
df5 = df5.drop(columns=['id', 'Unnamed: 32']) # menghapus kolom id dan unnamed
df5.head()
```

2. Lakukan proses encoding pada kolom "diagnosis".

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
```

- LabelEncoder: Digunakan untuk mengonversi variabel kategori menjadi angka, sehingga dapat digunakan dalam model machine learning.
- StandardScaler: Digunakan untuk menstandarkan fitur dengan mengurangi rata-rata dan membagi dengan deviasi standar, sehingga fitur memiliki distribusi normal.
- train_test_split: Fungsi untuk membagi dataset menjadi data pelatihan dan data pengujian, membantu dalam evaluasi model.

```python
le = LabelEncoder() # membuat objek dari LabelEncoder
df5['diagnosis'] = le.fit_transform(df5['diagnosis' ]) # proses encoding
```

```python
df5.head()
```

3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

```python
numerical = df5.columns[df5.columns != 'diagnosis']
scaler = StandardScaler()
df5[numerical] = scaler.fit_transform(df5[numerical])
df5.head()
```

- Baris pertama , Kode ini memilih semua kolom dalam DataFrame df, kecuali kolom 'diagnosis'. Variabel numerical berisi nama-nama kolom numerik.
df.columns: Mengambil daftar semua nama kolom yang terdapat dalam DataFrame df.
df.columns != 'diagnosis': Membuat kondisi boolean yang memeriksa apakah setiap nama kolom tidak sama dengan 'diagnosis'. Hasilnya adalah array boolean di mana nilai True menunjukkan kolom-kolom yang tidak dinamai 'diagnosis'.
df.columns[df.columns != 'diagnosis']: Menggunakan kondisi boolean tersebut untuk memilih nama-nama kolom yang tidak dinamai 'diagnosis', yaitu kolom-kolom numerik yang perlu distandarisasi.
- Baris kedua , Membuat objek StandardScaler untuk menstandarkan data.
- Baris ketiga , Menerapkan standarisasi pada kolom numerik dengan menghitung rata-rata dan deviasi standar, lalu mengubah data tersebut.
df[numerical]: Mengakses subset dari DataFrame df yang hanya mencakup kolom-kolom numerik (selain kolom 'diagnosis').
scaler.fit_transform(df[numerical]):
fit(): Menghitung rata-rata dan deviasi standar dari setiap fitur dalam kolom-kolom numerik.
transform(): Mengubah data dalam kolom-kolom numerik menggunakan rata-rata dan deviasi standar yang telah dihitung, sehingga setiap fitur memiliki rata-rata 0 dan deviasi standar 1.
Hasil dari fit_transform() adalah DataFrame yang berisi fitur-fitur numerik yang telah distandarisasi.
- Baris keempat , menampilkan data

4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
df5_train, df5_unsee = train_test_split(df5, test_size=0.2, random_state=0, stratify=df5['diagnosis'])
df5_val, df5_test = train_test_split(df5_unsee, test_size=0.5, random_state=0, stratify=df5_unsee['diagnosis'])

print(f'Jumlah data asli: {df5.shape[0]}')
print(f'Jumlah data train: {df5_train.shape[0]}')
print(f'Jumlah data unsee: {df5_unsee.shape[0]}')
print(f'Jumlah data val: {df5_val.shape[0]}')
print(f'Jumlah data test: {df5_test.shape[0]}')

# Cek rasio tiap label
print('=========')
print(f'Jumlah label data asli:\n{df5["diagnosis"].value_counts()}')
print(f'Jumlah label data train:\n{df5_train["diagnosis"].value_counts()}')
print(f'Jumlah label data val:\n{df5_val["diagnosis"].value_counts()}')
print(f'Jumlah label data test:\n{df5_test["diagnosis"].value_counts()}')
```

Baris pertama ,
- train_test_split: Fungsi ini digunakan untuk membagi dataset df5 menjadi dua bagian: df5_train (80% data) dan df5_unsee (20% data).
- test_size=0.2: Menentukan proporsi data yang akan digunakan untuk data yang tidak terlihat (unseen).
- random_state=0: Menjamin bahwa pembagian data bersifat reproducible (dapat diulang) pada setiap eksekusi.
- stratify=df5['diagnosis']: Menjaga proporsi label dalam kolom 'diagnosis' agar tetap sama di kedua subset.

Selanjutnya Baris kedua ,
- Kode ini membagi subset df5_unsee menjadi dua bagian yang sama: df5_val dan df5_test (masing-masing 50% dari df5_unsee).
Proses stratifikasi tetap diterapkan untuk menjaga proporsi label.

Baris 3-7 ,
Baris-baris ini mencetak jumlah total data pada setiap subset, yaitu data asli, data pelatihan, data unseen, data validasi, dan data pengujian.

Baris 8-akhir
Kode ini memeriksa dan mencetak jumlah setiap label dalam kolom 'diagnosis' untuk semua dataset (asli, train, val, dan test).
Fungsi value_counts() digunakan untuk menghitung frekuensi masing-masing label.
